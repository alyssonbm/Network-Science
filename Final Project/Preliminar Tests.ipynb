```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('Data/casos_coronavirus_2021_03_18_1.csv', sep=',', low_memory=False)
df2 = pd.read_csv('Data/casos_coronavirus_2021_03_18_2.csv', sep=',', low_memory=False)
df3 = pd.read_csv('Data/casos_coronavirus_2021_03_18_3.csv', sep=',', low_memory=False)
df4 = pd.read_csv('Data/casos_coronavirus_2021_03_18_4.csv', sep=',', low_memory=False)
df5 = pd.read_csv('Data/casos_coronavirus_2021_03_18_5.csv', sep=',', low_memory=False)
df6 = pd.read_csv('Data/casos_coronavirus_2021_03_18_6.csv', sep=',', low_memory=False)
df7 = pd.read_csv('Data/casos_coronavirus_2021_03_18_7.csv', sep=',', low_memory=False)
df8 = pd.read_csv('Data/casos_coronavirus_2021_03_18_8.csv', sep=',', low_memory=False)
df9 = pd.read_csv('Data/casos_coronavirus_2021_03_18_9.csv', sep=',', low_memory=False)
```
